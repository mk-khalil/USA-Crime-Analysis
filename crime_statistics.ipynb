## 1. National criminal offense rates per year across all available years for the top five most frequent offense categories

```python
''' transforming the state population dataframe form [state, 1991, 1992,..] to [state, year, population] '''
# transformed dataframe
df_popT = pd.DataFrame(columns=['state', 'year', 'population'])
# get the list of years from the original dataframe
years = df_pop.columns[1:]
# iterate through the rows of the original dataframe
for i, row in df_pop.iterrows():
    # get the state name
    state = row['state']
    
    # iterate through the years
    for year in years:
        # get the population for the current state and year
        population = row[year]
        
        # add a new row to the new dataframe with the state, year, and population
        df_popT = df_popT.append({'state': state, 'year': year, 'population': population}, ignore_index=True)

# changing columns type to int
df_popT['year'] = df_popT['year'].astype('int')
df_popT['population'] = df_popT['population'].astype('int')

# renaming the state column to merge the dataframes later
df_popT.rename(columns = {'state':'state_name'}, inplace = True)

# display the new dataframe
print(df_popT)
```

To calculate the yearly national criminal offense rates, we will need to divide the number of offenses for each category in each state in a given year by the state population in that year. This would give us the offense rate per 100,000 population for each category in each year for each state. 

Then for each year we could use the mean offense rate per 100,000 population over the offense rates of different states. This would give us the national offense rate for each category and year

The reason for following this analysis instead of summing the offense counts for all the states and dividing by the country's population is due to the fact that some criminal offense counts are not reported for some states in some years. So dividing by the whole population would lead to misleading rates.


To identify the top five most frequent offense categories, we could use the offense rates of each category and get their mean over the years. We could then tank the offense categories based on the mean offense rates and select the top five.


```python
# merging the offense and population dataframes on 'year' to get state offense rates
merged_df = pd.merge(df_nibrs, df_popT, on=['year', 'state_name'])
# Calculate the offense rate per 100,000 population
merged_df['offense_rate'] = merged_df['count'] / (merged_df['population'] / 100000)

# group the data by year and offense categores to get the mean of offense rates of all states in a given year
grouped_df = merged_df.groupby(['year', 'offense'])
# Calculate the mean offense rate per 100,000 population for each offense categories over the years
offense_rates_year = grouped_df['offense_rate'].mean().reset_index()
# the resultung dataframe has the national offense rates of each offense category per years for all years 
offense_rates_year
```

```python
# To get the five most frequent offenses

# Group the data by offense category
grouped_df = offense_rates_year.groupby(['offense'])
# Calculate the mean offense rate per 100,000 population for each offense categories over the years
offense_frequency = grouped_df['offense_rate'].mean().reset_index()
# rank the offense categories based on their mean rate
offense_frequency = offense_frequency.sort_values(by='offense_rate', ascending = False)
# a list for top 5 offenes
top5_offenses = offense_frequency['offense'].values.tolist()[:5]
print ('The five most frequent offenses and their offense rate per 100,000 population:\n', offense_frequency.head() )

# filter the yearly offense rates dataframe to select only the five most frequent offenses
offense_rates_top5 = offense_rates_year[offense_rates_year['offense'].isin(top5_offenses)]
```

## 2. The average percentage of violent crimes relative to total crime per state over all available years.


First, we will calculate the total number of crimes for each state for each year by summing up the offense counts for all crime categories.

Next, we will calculate the total number of violent crimes for each state for each year by summing up the offense counts for all violent crime categories.

Then, for each state, we will calculate the average percentage of violent crimes relative to total crime by dividing the total number of violent crimes by the total number of crimes and multiplying by 100.

```python
# a list of the violent crimes from the offense categories 
violent_crimes = ['assault-offenses', 'homicide-offenses', 'robbery','kidnapping-abduction','sex-offenses']

# Calculate the total number of crimes for each state for each year
df_total_crimes = df_nibrs.groupby(['state_name', 'year'])['count'].sum().reset_index()
# Calculate the total number of violent crimes for each state for each year
df_violent_crimes = df_nibrs[df_nibrs['offense'].isin(violent_crimes)].groupby(['state_name', 'year'])['count'].sum().reset_index()

# Merge the dataframes to get the total number of crimes and violent crimes for each state for each year
df_merged = pd.merge(df_total_crimes, df_violent_crimes, on=['state_name', 'year'])

# Calculate the average percentage of violent crimes for each state
df_avg = df_merged.groupby('state_name')[['count_x', 'count_y']].mean()
df_avg['percent_violent_crimes'] = df_avg['count_y'] / df_avg['count_x'] * 100
df_avg.drop(['count_x', 'count_y'], axis =1, inplace = True)

print ('The average precentage of violent crimes for all states = ',  mean(df_avg['percent_violent_crimes'].values.tolist()))
df_avg.head()
```

To visualize the results, it would be a good idea to create a bar plot showing the average percentage of violent crimes per state. This will allow us to easily compare the average percentage of violent crimes among the different states.

```python
# Sort the dataframe by the percentage of violent crimes
df_avg = df_avg.sort_values('percent_violent_crimes')

# Plot the results using a bar plot
plt.figure(figsize=(20, 15))
sns.barplot(x='state_name', y='percent_violent_crimes', data=df_avg.reset_index(),
            palette='rocket_r')
plt.xlabel('State', fontsize=18)
plt.ylabel('Percentage of violent crimes', fontsize=18)
plt.title('Average percentage of violent crimes per state', fontsize=18)
# Rotate the x-axis labels by 90 degrees
plt.xticks(rotation=90, fontsize=14)
plt.show()
```

**In this section, we will test the Claim that “U.S. states that implement stricter firearm control laws, have lower violent crime rates on average” using two-tailed t-test and linear regression analysis test**

### Two-tailed t-test
To assess the validity of the claim that "U.S. states that implement stricter firearm control laws have lower violent crime rates on average," we will use a two-tailed dependent samples t-test. This test is appropriate because we will be comparing the mean violent crime rates of states that have stricter firearm control laws to the mean violent crime rates of states that have less strict firearm control laws, and we have repeated measures (violent crime rates) for each state over the past 30 years.

The null hypothesis for the test is that there is no difference in the mean violent crime rates of states that have stricter firearm control laws versus those that have less strict laws. The alternative hypothesis is that there is a difference in the mean violent crime rates of states with stricter firearm control laws and those with less strict laws.

To conduct the test, we will need to first classify the states into two groups: those with stricter firearm control laws and those with less strict laws. we will do this by calculating the mean number of gun control laws applied in each state across all years in the dataset, and then dividing the states into two groups based on whether they have above or below the median number of laws applied.

Next, we will calculate the mean violent crime rate for each group of states. Finally, we will conduct the t-test by using a scipy statistical software package

```python
# Get the average violent crime offense rate over the years for each state

# merging the offense and population dataframes on 'year' and 'state_name' to get state offense rates
merged_df = pd.merge(df_nibrs, df_popT, on=['year', 'state_name'])
# Calculate the offense rate per 100,000 population
merged_df['offense_rate'] = merged_df['count'] / (merged_df['population'] / 100000)
# Select violent crime offenses 
merged_df = merged_df[merged_df['offense'].isin(violent_crimes)]


# group the data by year and states 
grouped_df = merged_df.groupby(['year', 'state_name'])
# Calculate the mean offense rate per 100,000 population for each offense category per state per year
violent_offense_rates_state_year = grouped_df['offense_rate'].mean().reset_index()

# get the mean of offense rates over the years for all states
violent_offenses_state = violent_offense_rates_state_year.groupby('state_name')['offense_rate'].mean().reset_index()
# rename states column for merging later
violent_offenses_state.rename(columns = {'state_name': 'state'}, inplace = True)

# the resultung dataframe has the average violent crime offense rates of all the states
violent_offenses_state.head()
```

A state gets a code 1 for a law if the state applies it and the law imposes more gun control or if the state doesn't apply it and the law faciltates carrying guns/firearms in any way and vice versa for code 0. For example, the preemption law states that State law does not preempt local regulation of firearms in any way. However the coding in such case is reversed so that any preemption of local regulation is coded as being preemption, and this provision would be coded as a 0.
```python
# Divide the states into two groups based on whether they have a mean number of gun control laws above or below the median
strict_gun_control_states = violence_gunlaws_df[violence_gunlaws_df['strict']]['offense_rate']
less_strict_gun_control_states = violence_gunlaws_df[ ~ (violence_gunlaws_df['strict'])]['offense_rate']

# Calculate the mean violent crime rate for each group of states
mean_violent_crime_strict_gun_control = strict_gun_control_states.mean()
mean_violent_crime_less_strict_gun_control = less_strict_gun_control_states.mean()

print('The mean violence crime rate of states with strict gun law = ', mean_violent_crime_strict_gun_control)
print('The mean violence crime rate of states with less strict gun law = ', mean_violent_crime_less_strict_gun_control )

print('-------------------------------------------------')
# Perform the t-test
t, p = stats.ttest_ind(less_strict_gun_control_states,strict_gun_control_states)


# Print the results
print("t-statistic:", t)
print("p-value:", p)
print('-------------------------------------------------')

# Check the p-value to see if the difference in means is statistically significant
if p < 0.05:
    print("The difference in mean violent crime rates between states with stricter firearm control laws and those with less strict laws is statistically significant.")
else:
    print("The difference in mean violent crime rates between states with stricter firearm control laws and those with less strict laws is not statistically significant.")
```

If the p-value is less than the chosen significance level of 0.05, we can reject the null hypothesis and conclude that there is a significant relationship between the number of gun control laws and the violent crime rate.
